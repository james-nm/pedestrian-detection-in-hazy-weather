```python
import os
```

```python
def recursive_listdir(path):
    path_list = []
    
    for item in os.listdir(path):
    
        if os.path.isdir(os.path.join(path, item)):
            
            path_list = path_list + recursive_listdir(os.path.join(path, item))
            
        else: 
            path_list.append(os.path.join(path, item))
            
    return path_list
    
```

```python
all_files = recursive_listdir('./')
print(all_files)
```

```python
for file_path in all_files:
    
    if file_path[-2:] == 'py':
        print ('updating ' + file_path)
        with open(file_path, 'r+') as file:

            file_source = file.read()
            replace_string = file_source.replace('slim = tf.contrib.slim', 'import tf_slim as slim')
            
        with open(file_path, 'r+') as file:
            file.write(replace_string)
            
            
    
```
